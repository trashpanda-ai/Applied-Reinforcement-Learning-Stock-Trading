```python
import tensorflow as tf
from sklearn.datasets import load_boston
import pylab as plt
import seaborn as sb
import numpy as np
import pandas as pd

tf.enable_eager_execution()


data = load_boston()

X = data.data
Y = data.target

if Y.ndim == 1:
    Y = Y[:,None]
mean_X = X.mean(axis=0)
std_X = X.std(axis=0)

mean_Y = Y.mean(axis=0)
std_Y = Y.std(axis=0)

X_n = (X - mean_X) / std_X
Y_n = (Y - mean_Y) / std_Y
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(10, activation='tanh', kernel_regularizer='l1'),
  tf.keras.layers.Dense(1, activation='linear', kernel_regularizer='l1')
])
model.compile(
    optimizer =tf.keras.optimizers.RMSprop(),
    #optimizer=tf.optimizers.Adam(learning_rate=0.01),
    #optimizer=tf.optimizers.RMSprop(),
    loss='mean_absolute_error'
)
model.fit(X_n, Y_n, epochs=3000)
```

```python
tf.enable_eager_execution()
xs = tf.Variable(X_n)
with tf.GradientTape() as tape:
    tape.watch(xs)
    pred = model(xs)
    print (pred)
    grad = tape.gradient(pred,xs)
print (xs.shape)
```

```python
fig, ax = plt.subplots(figsize=(15,4))
fig.suptitle("Absolute gradients on input layer over all pattern")
#cax = ax.matshow(abs(grad.numpy().T), aspect='auto')
cax = ax.matshow(abs(grad.numpy().T), aspect='auto')
plt.colorbar(cax)
x_pos = np.arange(len(data.feature_names))
plt.yticks(x_pos, data.feature_names)
plt.show()
```

```python
#sensitivity_inputLayer = np.sum (np.abs(grad), axis=0)
sensitivity_inputLayer_abs_sum = np.sum(np.abs(grad), axis=0)
df = pd.DataFrame ([data.feature_names, sensitivity_inputLayer_abs_sum])
df = df.transpose()
df.columns = ["feature", "sensitivity"]
df = df.sort_values("sensitivity")
df

plt.figure(figsize=(10,4))
plt.suptitle ("Sensitivity based on sum of absolute input layer gradients")
plt.bar(df.feature, df.sensitivity, align='center')
plt.grid()
plt.ylabel('Sensitivity')
plt.title('Boston Housing')
```

```python
sensitivity_inputLayer_mean = np.abs(np.mean(grad, axis=0))
sensitivity_inputLayer_std = np.std(grad, axis=0)

df = pd.DataFrame ([data.feature_names, sensitivity_inputLayer_mean,sensitivity_inputLayer_std])
df = df.transpose()
df.columns = ["feature", "sensitivity_mean", "sensitivity_std"]
df = df.sort_values("sensitivity_mean")
print (df)

x = np.arange(len(df.feature))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,4))
fig.suptitle ("Sensitivity based on abs(mean(grad)) of input layer gradients")
bar_mean = ax.bar(x - width/2, df.sensitivity_mean, width, label="mean")
bar_std  = ax.bar(x + width/2, df.sensitivity_std,  width, label="std")
ax.grid()
ax.legend(loc="best")
ax.set_ylabel('Sensitivity')
ax.set_xticks(x)
ax.set_xticklabels(df.feature)

plt.title('Boston Housing')
plt.savefig("sensitivities-inputLayer.png")
```
